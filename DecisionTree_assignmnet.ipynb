```python
#import the titanic data using a url
url = "https://raw.githubusercontent.com/BigDataGal/Python-for-Data-Science/master/titanic-train.csv"
titanic = pd.read_csv(url)
```

```python
#droping unnecessary columns
titanic_new = titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin', 'Embarked'], axis = 1)
```

# Data Cleaning

```python
#finding null values 
titanic_new.isnull().sum()
```

```python
#As "Sex" column has string type data like "male" and "female" so we have to convert into a numerical data
#here  we'll sklearn library preprocessing "LabelEncoder"
from sklearn.preprocessing import LabelEncoder
number = LabelEncoder()
titanic_new['Sex'] = number.fit_transform(titanic_new['Sex'].astype(str))
```

```python
titanic_new['Sex']
```

```python
#Also we have column called "Age" which has numeric bins value which will cause the error bcoz it is not able to comapre with other values
def imputate_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age)==1:
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
titanic_new["Age"] = titanic_new[['Age','Pclass']].apply(imputate_age,axis=1)
```

# Model Building

```python
# X independent variable and y dependent variable
X  = titanic_new.drop(columns = 'Survived')
y = titanic_new['Survived']
```

```python
#Split dataset into tarining and testing
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.30, random_state = 101)
```

```python
#visualize the data on tree
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
feature_name = list(X.columns)
class_name = list(y_train.unique())
feature_name
```

```python
class_name
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image 
#Create a dot file which stores the tree structure
dot_file = export_graphviz(classifier, feature_names = feature_name, rounded = True, filled = True)

#Draw the graph
graph = pydotplus.graph_from_dot_data(dot_file)
graph.write_png("MyTree.png")

#show the graph
Image(graph.create_png())
```

```python
classifier.score(X_train, y_train)
```

```python
pred = classifier.predict(X_test)
pred
```

```python
#accuracy of our classification tree
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
Acuuracy = accuracy_score( y_test, pred)
Acuuracy
```

```python
con_matrix = confusion_matrix(y_test,pred)
con_matrix
```

```python
print(f"Cf_report : {classification_report(y_test,pred)}")
```

# Apply hyperperameter for increasing accuracy

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
X_transform = scalar.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_transform, y, test_size = 0.30, random_state=100)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(X_transform)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.show()
```

As we can see 95% percent varience is being explained by 5 components, so we will give only these 5 components to increase accuracy

```python
pca = PCA(n_components = 5)
new_data = pca.fit_transform(X_transform)

principal_X = pd.DataFrame(new_data,columns=['PC-1','PC-2','PC-3','PC-4','PC-5'])
```

```python
principal_X
```

```python
#now we'll see how much accuracy is increased or not
X_train, X_test, y_train, y_test = train_test_split(principal_X, y, test_size = 0.30, random_state = 101)
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

As we can see accuracy is increased

Now try to tune our model with GridSearchCV.

```python
# we are tuning three hyperparameters right now, we are passing the different values for both parameters
grid_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,32,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'splitter' : ['best', 'random']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=clf,
                     param_grid=grid_param,
                     cv=5,
                    n_jobs =-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
print(best_parameters)
```

```python
grid_search.best_score_
```
